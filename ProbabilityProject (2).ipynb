Displaying the top 10 artists

```python
all_years = df[df['released_year'] > 1987]
top_artists = all_years['artist(s)_name'].value_counts().head(10)
print(top_artists)
```

Top 10 Artists by Number of Songs Released (1987-2023)

```python
plt.figure(figsize=(10, 6))
sns.barplot(y=top_artists.index, x=top_artists.values)
plt.title('Top 10 Artists by Number of Songs Released (1987-2023)')
plt.ylabel('Artist')
plt.xlabel('Number of Songs Released')
plt.xticks(rotation=45)
plt.show()
```

```python
df["key"].value_counts().count()
```

```python
sns.barplot(y="key", x=df.key.index, data=df).set_title("Key Variable Frequencies");
```

```python
df["mode"].value_counts().plot.bar().set_title("Mode Variable Frequencies");
```

Mode Distribution Percentage Pie Chart

```python
labels=["Major","minor"]

size=df["mode"].value_counts()

colors=plt.cm.Wistia(np.linspace(0,1,2))

explode=[0,0.05]

plt.rcParams["figure.figsize"]=(5,5)

plt.pie(size, labels=labels,colors=colors, explode=explode, shadow = True, startangle=90,autopct='%1.1f%%')

plt.title("Mode Distribution Percentage", fontsize = 15)

plt.legend()

plt.show()
```

Relationship between Key and Mode with by chart

```python
plt.figure(figsize=(10, 6))
df.groupby('key')['mode'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.title('Relationship between Key and Mode')
plt.xlabel('Key')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Plot distributions for numerical features

```python

numerical_columns = [ 'bpm','danceability_%', 'valence_%',
                     'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']

plt.figure(figsize=(15, 20))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(5, 4, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

Filter DataFrame by numeric columns with plot correlation heatmap

```python
numerical_columns = [ 'streams', 'in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists', 'in_apple_charts',
                     'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts','bpm','danceability_%', 'valence_%',
                     'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']

numeric_df = df[numerical_columns]

plt.figure(figsize=(12, 10))
correlation_matrix = numeric_df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,cbar=False)
plt.title('Correlation Heatmap')
plt.show()
```

Distribution of Danceability, Energy, and Valence

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=df[['danceability_%', 'energy_%', 'valence_%']], inner='points')
plt.title('Distribution of Danceability, Energy, and Valence')
plt.xlabel('Feature')
plt.ylabel('Value')
plt.show()
```

Checking outliers in numerical variables with boxplot

```python
numerical_columns = [ 'bpm','danceability_%', 'valence_%',
                     'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```
